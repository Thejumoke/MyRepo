```python
# Create list years
years = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
```

```python
# Create list durations
durations = [103,101,99,100,100,95,95,96,93,90]
```

```python
# Create dictionary movie_dict
movie_dict = {'years':[2011,2012,2013,2014,2015,2016,2017,2018,2019,2020], 'durations':[103,101,99,100,100,95,95,96,93,90]}
```

```python
# Print movie_dict
print(movie_dict)
```

```python
import pandas as pd
```

```python
# Create dataframe durations_df from dictionary movie_dict
durations_df =pd.DataFrame(movie_dict)
```

```python
# Print first five rows of dataframe durations_df
print(durations_df)

```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
# Create line plot of data with years in x-axis and durations in y-axis, add title to the plot
plt.plot(durations_df["years"], durations_df["durations"])
plt.title("Netflix Movie Durations 2011-2020")
plt.xlabel('years')
plt.ylabel('durations')
plt.show()
```

```python
#Import data from CSV and print first 5 rows
netflix_df = pd.read_csv('C:/Users/hp/Documents/CWW Africa/CWW Tech 4.0 Individual Project/datasets/netflix_data.csv')
print(netflix_df.head())
```

```python
# Subset dataframe such that only rows where type is 'Movie are preserved
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']
```

```python
# Subset dataframe such that only columns title, country, genre, release_date and duration are preserved
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]
# Print result
print(netflix_movies_col_subset.head())
```

```python
#Create scatter plot for release_year and duration and include title
fig = plt.figure(figsize=(12,8))
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'])
plt.xlabel('release_year')
plt.ylabel('duration')
plt.title('Movie Duration by Year of Release')
plt.show()
```

```python
# Subset Dataframe and preserve movies with duration less than 60 minutes
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] <60]
print(short_movies.head(20))
```

```python
#Initialize empty list
colors =[]
```

```python
for lab, row in netflix_movies_col_subset.iterrows() :
    if row['genre'] == 'Children' :
        colors.append('red')
    elif row['genre'] == 'Documentaries' :
        colors.append('blue')
    elif row['genre'] == 'Stand-Up' :
        colors.append('green')
    else:
        colors.append('black')
        
      
```

```python
print(colors[:10])
```

```python

#Set figure style and initiate new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))

# Create scatter plot with release year and duration
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], c=colors)

#Add labels and title
plt.xlabel('Release year')
plt.ylabel('Duration(min)')
plt.title('Movie duration by year of release')

#Show plot
plt.show()
```
